```python
# Coloca tus credenciales aquí
CLIENT_ID = input('pon tu id de api spotify')
CLIENT_SECRET= input('pon tu secreto')
# Autenticación
auth_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
datos_2024 = sp.search(q = "year:2024", limit=5, offset=0, type='track', market="ES")
datos_2024
```
