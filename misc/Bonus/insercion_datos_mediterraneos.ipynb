```python
mycursor = cnx.cursor()
```

```python
query = """
INSERT INTO canciones (id_cancion, nombre_cancion, popularity, artista, id_pais, genero, rango_lustros)
VALUES (%s, %s, %s, %s, %s, %s, %s)"""
```

```python
for _, row in df.iterrows():
    values = (
        row['id'],
        row['nombre_cancion'], 
        row['popularity'], 
        row['artista'], 
        row['pais'], 
        row['genero'],
        row['Rango_lustros']
    )
    try:
        mycursor.execute(query, values)
    except mysql.connector.Error as err:
        print(f"Error: {err}")
        cnx.rollback()
        continue

    cnx.commit()   
```
