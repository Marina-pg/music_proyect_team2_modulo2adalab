```python
import requests
import pandas as pd
```

```python
api_key = input('inserta aqui tu api key')
```

```python
requests.get(f'http://www.last.fm/api/auth/?api_key={api_key}')
```

```python
genero = input('inserta el género que quieres explorar')
```

```python
top_tracks_call= requests.get(f'https://ws.audioscrobbler.com/2.0/?method=tag.gettoptracks&tag={genero}&api_key={api_key}&format=json')
```

```python
top_tracks_call
```

```python
top_tracks = top_tracks_call.json()
```

```python
top_tracks
```

```python
top_10 = []
for track in top_tracks['tracks']['track'][0:10]:
    del track['mbid']
    del track['streamable']
    del track['artist']['mbid']
    del track['artist']['url']
    top_10.append(track)
print(top_10) # Verifica que la estructura se haya limpiado
```

```python
for track in top_10[0:10]:
    del track['url']
    del track['image']
print(top_10)
```

```python
pd.json_normalize(top_10)
```
