```python
import requests
import pandas as pd
```

```python
api_key = input ('inserta tu api key')
```

```python
pais = input('Indica el pa√≠s')
```

```python
track_geo_call = requests.get(f'https://ws.audioscrobbler.com/2.0/?method=geo.gettoptracks&country={pais}&api_key={api_key}&format=json')
```

```python
top_tracks_geo = track_geo_call.json()
```

```python
top_tracks_geo
```

```python
top_5 = []
for track in top_tracks_geo['tracks']['track'][0:10]:
    del track['mbid']
    del track['streamable']
    del track['artist']['mbid']
    del track['artist']['url']
    del track['url']
    del track['image']
    top_5.append(track)
print(top_5) # Verifica que la estructura se haya limpiado
```

```python
pd.json_normalize(top_5)
```
