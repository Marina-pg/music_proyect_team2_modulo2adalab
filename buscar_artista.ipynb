```python
import requests
import pandas as pd
```

```python
api_key = input('inserta tu apikey de last.fm')
```

```python
artist = input('Indica el artista')
```

```python
top_tracks_call= requests.get(f'https://ws.audioscrobbler.com/2.0/?method=artist.search&artist={artist}&api_key={api_key}&format=json')
```

```python
top_tracks_call.json()
```

```python
top_tracks = top_tracks_call.json()
```

```python
top_10 = []
for track in top_tracks['tracks']['track'][0:10]:
    del track['mbid']
    del track['streamable']
    del track['artist']['mbid']
    del track['artist']['url']
    del track['url']
    del track['image']
    top_10.append(track)
print(top_10) # Verifica que la estructura se haya limpiado
```

```python
pd.json_normalize(top_10)
```
